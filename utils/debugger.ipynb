```python
def crc32(word: int) -> int:    # calcola il CRC di una riga da 32 bit del pacchetto
    step_1 = (word >> 16) ^ (word & 0xFFFF)
    step_2 = (step_1 >> 8) ^ (step_1 & 0xFF)
    return (step_2 >> 4) ^ (step_2 & 0xF)

def crc32check(arr) -> str:     # prende in input un array contenente le varie righe del
    testvalue = (arr[-1] & 0xF)   # pacchetto e verifica il CRC
    crc = crc32(arr[0])
    for i in range(1, len(arr)):
        print(crc, 'xor')
        if i != len(arr)-1:
            crc = crc32(arr[i]) ^ crc
        else:
            crc = crc32(arr[i] & 0xFFFFFFF0) ^ crc
    if crc == testvalue:
        print(crc, '=', testvalue)
        return 'Correct'
    else:
        print(crc, '=/=', testvalue)
        return 'Transmission error'

value = [0b10000000001000010010111001000000,
0b00011000011110011110100011000110,
0b01000000001001011000100111001100,
0b01000000010000010011001011101001,
0b11000100110100000000000000011001]
print(crc32check(value))

print(crc32(0x80212e40))
```
